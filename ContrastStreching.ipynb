```python
#import library
import numpy #-> library perhitungan
import matplotlib.pyplot as plt #-> Library untuk buat grafik
from copy import deepcopy #library untuk cloning atau copy
from PIL import Image #->ngambil ngambar
```

```python
#function untuk balikin nilai RGB
def getGrayColor(rgb):
  return rgb[0]
```

```python
#Balikin nilai color dalam index RGB
def setGrayColor(color):
  return [color, color, color]
```

```python
#image open
img = Image.open("./MartinelliBNW.jpg")
img = numpy.asarray(img)
#clonng image
ct = deepcopy(img)

#nilai awal
r1 = 100
s1 = 50
r2 = 150
s2 = 200
```

```python
#nested loop -> contrast stretching
for i in range(len(img)):
  for j in range(len(img[i])):
    x = getGrayColor(img[i][j])
    if(0 <= x and x <= r1):
      ct[i][j] = setGrayColor(s1/r1 * x)
    elif(r1 < x and x <= r2):
      ct[i][j] = setGrayColor(((s2 - s1) / (r2 - r1)) * (x - r1) + s1)
    elif(r2 < x and x <= 255):
      ct[i][j] = setGrayColor(((255 - s2) / (255 - r2)) * (x - r2) + s2)
```

```python
#image lama/ original image
plt.subplot(2, 2, 1)
plt.imshow(img)
#image setelah dia strech
plt.subplot(2, 2, 2)
plt.imshow(ct)
plt.show()
```
